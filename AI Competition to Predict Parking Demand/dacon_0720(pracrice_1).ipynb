### 10분내지하철수, 10분내버스정류장수, 공급유형 결측치 처리후 모델생성

```python
from sklearn.model_selection import train_test_split
```

```python
sel = [ '총세대수', '전용면적별세대수', '10분내버스정류장수', '단지내주차면수','10분내지하철수']

X = train[sel]
y = train['등록차량수']
test_X = test[sel]

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.1, 
                                                    random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("학습(score) :", model.score(X_train, y_train) ) # 결정계수
print("테스트(score) :", model.score(X_test, y_test) ) # 결정계수
```

```python
mae_val = np.mean( abs( y_test - pred ) )
print( mae_val )
mse_val = np.mean( (y_test - pred) **2 )
print( mae_val )
rmse_val = mse_val ** 0.5
print( rmse_val )
```

```python
model = RandomForestRegressor(n_jobs=-1)
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("학습(score) :", model.score(X_train, y_train) ) # 결정계수
print("테스트(score) :", model.score(X_test, y_test) ) # 결정계수
```

```python
mae_val = np.mean( abs( y_test - pred ) )
print( mae_val )
mse_val = np.mean( (y_test - pred) **2 )
print( mae_val )
rmse_val = mse_val ** 0.5
print( rmse_val )
```

```python
import xgboost as xgb

data_dmatrix = xgb.DMatrix(data=X_train,label=y_train)
# 기본 옵션 확인
xg_reg = xgb.XGBRegressor(objective ='reg:linear', 
            colsample_bytree = 0.33, # 각나무마다 사용하는 feature 비율
            learning_rate = 0.1,
            max_depth = 5, 
            alpha = 0.1, 
            n_estimators = 2500)  # n_estimators=2500

xg_reg.fit(X_train, y_train)
pred = xg_reg.predict(X_test)

print("학습(score) :", xg_reg.score(X_train, y_train) ) # 결정계수
print("테스트(score) :", xg_reg.score(X_test, y_test) ) # 결정계수
```
